```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data.csv", encoding="latin1")

print("Columns in dataset:")
print(df.columns)

x = df["no2"].dropna().astype(float).values

r = 102303323

ar = 0.05 * (r % 7)      # = 0.05
br = 0.3 * (r % 5 + 1)   # = 1.2

z = x + ar * np.sin(br * x)

mu_hat = np.mean(z)
sigma2_hat = np.var(z)

lam_hat = 1 / (2 * sigma2_hat)
c_hat = 1 / np.sqrt(2 * np.pi * sigma2_hat)

print("\n===== Estimated PDF Parameters =====")
print(f"λ (lambda) = {lam_hat}")
print(f"μ (mu)     = {mu_hat}")
print(f"c          = {c_hat}")

def pdf(z, lam, mu, c):
    return c * np.exp(-lam * (z - mu)**2)

z_sorted = np.sort(z)
pz = pdf(z_sorted, lam_hat, mu_hat, c_hat)

plt.figure(figsize=(8, 5))
plt.hist(z, bins=50, density=True, alpha=0.5, label="Empirical z")
plt.plot(z_sorted, pz, color="red", linewidth=2, label="Learned PDF")
plt.xlabel("z")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Learned Probability Density Function")
plt.show()
```
